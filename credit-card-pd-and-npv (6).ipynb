STEP 1 --- SQL

```python
# ═══════════════════════════════════════════════════════════
# SQL
# ───────────────────────────────────────────────────────────


import os
import pandas as pd
import numpy as np
import gc

print("1. Locating dataset...")
target_file = 'accepted_2007_to_2018Q4.csv'
actual_path = None

for dirname, _, filenames in os.walk('/kaggle/input'):
    if target_file in filenames:
        actual_path = os.path.join(dirname, target_file)
        break

if actual_path is None:
    raise FileNotFoundError("Could not find the dataset.")

print(f"File found at: {actual_path}")
print("2. Loading dataset into memory using PyArrow")


df = pd.read_csv(actual_path, engine='pyarrow')

print(f"Raw data loaded. Initial shape: {df.shape[0]:,} rows | {df.shape[1]} columns")

print("3. Filtering for closed loans and creating target variable")
valid_statuses = ['Fully Paid', 'Charged Off']
df = df[df['loan_status'].isin(valid_statuses)].copy()
df['target'] = np.where(df['loan_status'] == 'Charged Off', 1, 0)

print("4. Dropping post-outcome 'leaky' variables")
leaky_cols = [
    'recoveries', 'collection_recovery_fee', 'total_rec_prncp', 
    'last_pymnt_d', 'out_prncp', 'total_pymnt', 'total_rec_int', 
    'next_pymnt_d', 'last_credit_pull_d'
]
df = df.drop(columns=[col for col in leaky_cols if col in df.columns])

print("5. Dropping sparse columns (>50% missing data)")
missing_fractions = df.isnull().mean()
cols_to_keep = missing_fractions[missing_fractions <= 0.50].index
df = df[cols_to_keep]

gc.collect()

print("\n" + "="*50)
print(f"STEP 1 COMPLETE!")
print(f"Final Cleaned Data Shape: {df.shape[0]:,} rows | {df.shape[1]} columns")
print("="*50)
```

STEP 2 --- Time Based Split

```python
# ═══════════════════════════════════════════════════════════
# TIME BASED SPLIT
# ───────────────────────────────────────────────────────────


import pandas as pd
import gc

print("1. Converting dates")
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%Y')

print("2. Slicing data chronologically into Train and Test sets")
# Train Set: Everything before January 2016
train_df = df[df['issue_d'] < '2016-01-01'].copy()

# Test Set: Everything from January 2016 onwards
test_df = df[df['issue_d'] >= '2016-01-01'].copy()

print("3. Clearing original dataframe from RAM.")
#  destroy the original 1.3 million row dataset
del df
gc.collect()

print("4. Removing 'issue_d' from the training set to prevent temporal overfitting...")
# We keep issue_d in the test set to merge FRED macroeconomic rates later, but the XGBoost model shouldn't train on it.
train_df = train_df.drop(columns=['issue_d'])

print("\n" + "="*50)
print("STEP 2 COMPLETE: TIME-BASED SPLIT")
print(f"Training Set (2007-2015): {train_df.shape[0]:,} rows")
print(f"Testing Set (2016-2018):  {test_df.shape[0]:,} rows")
print("RAM successfully cleared and secured.")
print("="*50)
```

STEP 3 - Feature Engineering

 WITH 11 features

```python
# ═══════════════════════════════════════════════════════════
# FEATURE ENGINEERING
# ───────────────────────────────────────────────────────────



import pandas as pd
import numpy as np
import gc
import re

print("=" * 60)
print("STEP 3: FEATURE ENGINEERING & PIPELINE ARCHITECTURE")
print("=" * 60)

# ─────────────────────────────────────────────────────────────
# PHASE 1: TARGET ISOLATION
# ─────────────────────────────────────────────────────────────
print("\n[1/7] Isolating target variable...")

y_train = train_df['target'].astype('int8')
y_test  = test_df['target'].astype('int8')

print(f"      Train default rate : {y_train.mean():.2%}")
print(f"      Test  default rate : {y_test.mean():.2%}")

# ─────────────────────────────────────────────────────────────
# PHASE 2: NUCLEAR LEAKAGE FIREWALL
# ─────────────────────────────────────────────────────────────
print("\n[2/7] Applying leakage firewall")

leaky_columns = [
    # Post-outcome payment records
    'loan_status', 'out_prncp', 'out_prncp_inv',
    'total_pymnt', 'total_pymnt_inv', 'total_rec_prncp',
    'total_rec_int', 'total_rec_late_fee',
    'recoveries', 'collection_recovery_fee',
    'last_pymnt_d', 'last_pymnt_amnt', 'next_pymnt_d',
    'last_credit_pull_d',

    # Funded amounts (known only after LC approves — circular)
    'funded_amnt', 'funded_amnt_inv',

    # LC's own risk grade/rate — these ARE the bank's model output.
    # Removing them forces the model to learn default risk independently.
    'grade', 'sub_grade', 'int_rate', 'installment',

    # FICO at last pull (only known after loan is live — data leakage)
    'last_fico_range_high', 'last_fico_range_low',

    # Hardship / settlement flags (only set after borrower struggles)
    'hardship_flag', 'hardship_type', 'hardship_reason',
    'hardship_status', 'deferral_term', 'hardship_amount',
    'hardship_start_date', 'hardship_end_date',
    'payment_plan_start_date', 'hardship_length', 'hardship_dpd',
    'hardship_loan_status', 'orig_projected_additional_accrued_interest',
    'hardship_payoff_balance_amount', 'hardship_last_payment_amount',
    'debt_settlement_flag', 'debt_settlement_flag_date',
    'settlement_status', 'settlement_date', 'settlement_amount',
    'settlement_percentage', 'settlement_term',

   # Target itself
    'target', 'issue_d', 'pymnt_plan',
]

cols_to_drop_train = [c for c in leaky_columns if c in train_df.columns]
cols_to_drop_test  = [c for c in leaky_columns if c in test_df.columns]

X_train = train_df.drop(columns=cols_to_drop_train)
X_test  = test_df.drop(columns=cols_to_drop_test)

print(f"      Dropped {len(cols_to_drop_train)} leaky columns from train")
print(f"      X_train shape: {X_train.shape}")

# ─────────────────────────────────────────────────────────────
# PHASE 3: NUMERICAL IMPUTATION
# Fill missing numeric values with training-set medians.
# We compute medians ONLY on train to prevent test-set leakage.
# ─────────────────────────────────────────────────────────────
print("\n[3/7] Imputing missing values with train medians")

num_cols      = X_train.select_dtypes(include=[np.number]).columns
train_medians = X_train[num_cols].median()

X_train[num_cols] = X_train[num_cols].fillna(train_medians)
X_test[num_cols]  = X_test[num_cols].fillna(train_medians)  # use TRAIN medians on test

print(f"      Imputed {len(num_cols)} numerical columns")

# ─────────────────────────────────────────────────────────────
# PHASE 4: FEATURE ENGINEERING
# All features computed from information available at loan
# origination time. No future information used.
# ─────────────────────────────────────────────────────────────
print("\n[4/7] Engineering predictive features")

engineered = []


# 1. Loan-to-Income Ratio
#    Affordability: how big is this loan relative to annual income
#    High ratio → borrower is over-extended
if 'loan_amnt' in num_cols and 'annual_inc' in num_cols:
    X_train['loan_to_income'] = X_train['loan_amnt'] / (X_train['annual_inc'] + 1)
    X_test['loan_to_income']  = X_test['loan_amnt']  / (X_test['annual_inc']  + 1)
    engineered.append('loan_to_income')

# 2. Payment Burden — monthly instalment as % of monthly income
#    Captures cash-flow squeeze. >30% is considered high-risk in banking.
if 'installment' in train_df.columns and 'annual_inc' in num_cols:
    train_install = train_df['installment'].reindex(X_train.index).fillna(train_medians.get('installment', 0))
    test_install  = test_df['installment'].reindex(X_test.index).fillna(train_medians.get('installment', 0))
    X_train['payment_burden'] = train_install / (X_train['annual_inc'] / 12 + 1)
    X_test['payment_burden']  = test_install  / (X_test['annual_inc']  / 12 + 1)
    engineered.append('payment_burden')

# 3. FICO Midpoint — average of low/high FICO band (cleaner single score)
if 'fico_range_low' in num_cols and 'fico_range_high' in num_cols:
    X_train['fico_mid'] = (X_train['fico_range_low'] + X_train['fico_range_high']) / 2
    X_test['fico_mid']  = (X_test['fico_range_low']  + X_test['fico_range_high'])  / 2
    engineered.append('fico_mid')

# 4. Rate Above Cap — policy-specific: how far above 10% is this loan?
#    0 for loans already at/below 10%; positive for loans that would be banned.
if 'int_rate' in test_df.columns:
    test_df['rate_above_cap'] = np.maximum(test_df['int_rate'] - 10.0, 0)
    # Mirror into X_test so model can learn cap-related patterns
    X_test['rate_above_cap'] = test_df['rate_above_cap'].reindex(X_test.index).fillna(0)
    X_train['rate_above_cap'] = 0   # cap didn't exist at origination; set 0 for training
    engineered.append('rate_above_cap')

# 5. Repayment Ratio — how much of the principal has been paid back?

if 'total_pymnt' in test_df.columns:
    test_df['repayment_ratio'] = test_df['total_pymnt'] / (test_df['loan_amnt'] + 1)

# 6. Debt Service Ratio — annual debt payments as % of annual income
#    Unlike payment_burden (monthly), this captures the full-year picture.

if 'installment' in train_df.columns and 'annual_inc' in num_cols:
    X_train['debt_service_ratio'] = (train_install * 12) / (X_train['annual_inc'] + 1)
    X_test['debt_service_ratio']  = (test_install  * 12) / (X_test['annual_inc']  + 1)
    engineered.append('debt_service_ratio')

# 7. Revolving Utilisation Rate — normalised 0 to 1
#    High revol_util = borrower is maxed out on existing credit lines.
if 'revol_util' in num_cols:
    X_train['revol_utilisation_rate'] = X_train['revol_util'] / 100
    X_test['revol_utilisation_rate']  = X_test['revol_util']  / 100
    engineered.append('revol_utilisation_rate')

# 8. Delinquency Density — past defaults normalised by open accounts
#    2 delinquencies across 20 accounts (careful borrower) ≠
#    2 delinquencies across 3 accounts (serial defaulter).
if 'delinq_2yrs' in num_cols and 'open_acc' in num_cols:
    X_train['delinquency_density'] = X_train['delinq_2yrs'] / (X_train['open_acc'] + 1)
    X_test['delinquency_density']  = X_test['delinq_2yrs']  / (X_test['open_acc']  + 1)
    engineered.append('delinquency_density')

# 9. Long-Term Loan Flag — binary: 1 if 60-month loan, 0 if 36-month
#    60-month loans default ~30% more than 36-month in LC data.
#    Longer exposure window = more life events that can trigger default.
if 'term' in train_df.columns:
    train_term = train_df['term'].astype(str).str.extract(r'(\d+)').astype(float).squeeze()
    test_term  = test_df['term'].astype(str).str.extract(r'(\d+)').astype(float).squeeze()
    X_train['long_term_flag'] = (train_term.reindex(X_train.index).fillna(36) == 60).astype(int)
    X_test['long_term_flag']  = (test_term.reindex(X_test.index).fillna(36)  == 60).astype(int)
    engineered.append('long_term_flag')

# 10. Interest Cost Ratio — total lifetime interest / principal
#     High ratio = borrower is paying mostly interest, very little principal.
#     Signals borrower is reaching for cash they can barely afford.
#     Formula: (monthly_payment × term_months - principal) / principal
if 'installment' in train_df.columns and 'term' in train_df.columns:
    train_term_months = train_df['term'].astype(str).str.extract(r'(\d+)').astype(float).squeeze()
    test_term_months  = test_df['term'].astype(str).str.extract(r'(\d+)').astype(float).squeeze()

    train_interest = (train_install * train_term_months.reindex(X_train.index).fillna(36)) - train_df['loan_amnt'].reindex(X_train.index).fillna(0)
    test_interest  = (test_install  * test_term_months.reindex(X_test.index).fillna(36))   - test_df['loan_amnt'].reindex(X_test.index).fillna(0)

    X_train['interest_cost_ratio'] = train_interest / (train_df['loan_amnt'].reindex(X_train.index).fillna(1) + 1)
    X_test['interest_cost_ratio']  = test_interest  / (test_df['loan_amnt'].reindex(X_test.index).fillna(1)  + 1)
    engineered.append('interest_cost_ratio')

# 11. Cap Revenue Loss % — policy simulation feature
#     What % of a loan's interest revenue is destroyed by the 10% cap?
#     0% = loan already charges ≤10% (cap has no effect)
#     90% = loan charges 100% but cap allows only 10% (massive compression)
if 'int_rate' in test_df.columns:
    test_df['cap_revenue_loss_pct'] = (
        np.maximum(test_df['int_rate'] - 10.0, 0) / (test_df['int_rate'] + 1e-6)
    )
    # Mirror into X_test for model to learn margin-compression patterns
    X_test['cap_revenue_loss_pct']  = test_df['cap_revenue_loss_pct'].reindex(X_test.index).fillna(0)
    X_train['cap_revenue_loss_pct'] = 0
    engineered.append('cap_revenue_loss_pct')

# 12. Loan-to-Revolving Balance Ratio — debt consolidation signal
#     Ratio > 1 means the new loan is LARGER than total revolving debt.
#     sign of a debt-spiral: borrower keeps rolling debt into
#     larger and larger personal loans to keep afloat.
if 'loan_amnt' in num_cols and 'revol_bal' in num_cols:
    X_train['loan_to_revolving_ratio'] = X_train['loan_amnt'] / (X_train['revol_bal'] + 1)
    X_test['loan_to_revolving_ratio']  = X_test['loan_amnt']  / (X_test['revol_bal']  + 1)
    engineered.append('loan_to_revolving_ratio')

print(f"      Created {len(engineered)} engineered features:")
for f in engineered:
    print(f"        + {f}")

# ─────────────────────────────────────────────────────────────
# PHASE 5: CATEGORICAL ENCODING
# Drop high-cardinality noise, one-hot encode the rest.
# ─────────────────────────────────────────────────────────────
print("\n[5/7] Encoding categorical features")

cat_cols = X_train.select_dtypes(exclude=[np.number]).columns

# Drop high-cardinality columns (zip codes, free-text fields, etc.)
# These add memory cost and noise without signal.
high_cardinality = [c for c in cat_cols if X_train[c].nunique() > 50]
X_train = X_train.drop(columns=high_cardinality)
X_test  = X_test.drop(columns=high_cardinality)
print(f"      Dropped {len(high_cardinality)} high-cardinality columns")

# One-hot encode remaining categoricals
safe_cats = [c for c in cat_cols if c not in high_cardinality]
for col in safe_cats:
    X_train[col] = X_train[col].astype(str)
    X_test[col]  = X_test[col].astype(str)

X_train_encoded = pd.get_dummies(X_train, columns=safe_cats, drop_first=True, dtype='int8')
X_test_encoded  = pd.get_dummies(X_test,  columns=safe_cats, drop_first=True, dtype='int8')

# Align columns — test must have exactly the same columns as train
X_train_encoded, X_test_encoded = X_train_encoded.align(
    X_test_encoded, join='left', axis=1, fill_value=0
)
print(f"      Encoded {len(safe_cats)} categorical columns")

# ─────────────────────────────────────────────────────────────
# PHASE 6: XGBOOST COLUMN NAME SANITISATION
# ─────────────────────────────────────────────────────────────
print("\n[6/7] Sanitising column names for XGBoost")

def clean_column_names(df):
    df.columns = [re.sub(r'[\[\]<]', '_', str(col)) for col in df.columns]
    return df

X_train_encoded = clean_column_names(X_train_encoded)
X_test_encoded  = clean_column_names(X_test_encoded)

# ─────────────────────────────────────────────────────────────
# PHASE 7: MEMORY MANAGEMENT
# ─────────────────────────────────────────────────────────────
print("\n[7/7] Freeing RAM...")

del train_df, X_train, X_test
gc.collect()

# ─────────────────────────────────────────────────────────────
# SUMMARY
# ─────────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("STEP 3 COMPLETE")
print("=" * 60)
print(f"  X_train_encoded : {X_train_encoded.shape[0]:>10,} rows × {X_train_encoded.shape[1]} features")
print(f"  X_test_encoded  : {X_test_encoded.shape[0]:>10,} rows × {X_test_encoded.shape[1]} features")
print(f"  Engineered feats: {len(engineered)}")
print(f"  test_df retained: YES (for Step 5 profitability simulation)")
print("=" * 60)
```

STEP 4 --- RANDOM SEARCH

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import gc

print("="*60)
print("CONTINUOUS HYPERPARAMETER SEARCH")
print("="*60)

 
print("\n0. Calculating Class Imbalance Ratio")
ratio = float((y_train == 0).sum()) / (y_train == 1).sum()
print(f"Ratio secured: {ratio:.2f} (Safe loans for every 1 Default)")

print("\n1. Defining Mathematical Boundaries")

param_distributions = {
    # Search ANY floating decimal between 0.01 and 0.30
    'learning_rate': stats.uniform(loc=0.01, scale=0.29), 
    
    # Search ANY whole integer from 3 up to (but not including) 9
    'max_depth': stats.randint(low=3, high=9),
    
    'n_estimators': [100] # Keeping trees fixed to save time
}

print("Allowing the AI to hunt for the optimal combination")

# 2. Initialize the Scikit-Learn wrapper for XGBoost
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    scale_pos_weight=ratio, 
    tree_method='hist',
    random_state=42,
    n_jobs=2 
)

# 3. Setup the Randomized Search

random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_distributions,
    n_iter=15, 
    scoring='roc_auc',
    cv=3, 
    verbose=1,
    random_state=42
)

print("\n2. Initiating Continuous Search")
random_search.fit(X_train_encoded, y_train)

print("\n" + "="*60)
print("SEARCH COMPLETE: BEST PARAMETERS SECURED")
print("="*60)

# Extract the absolute best combination
best_params = random_search.best_params_
best_score = random_search.best_score_

print(f"Winning Learning Rate: {best_params['learning_rate']:.5f}")
print(f"Winning Max Depth:     {best_params['max_depth']}")
print(f"Highest Peak ROC-AUC:  {best_score:.4f}")
print("="*60)

print("\n3. Generating Performance Scatter Plot")
# Extract results for the graph
results = pd.DataFrame(random_search.cv_results_)

# Plotting the results
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    x=results['param_learning_rate'], 
    y=results['param_max_depth'], 
    c=results['mean_test_score'], 
    cmap='viridis', 
    s=150, # Size of the dots
    edgecolor='black',
    alpha=0.8
)

# Add a colorbar to show the AUC score mapping
cbar = plt.colorbar(scatter)
cbar.set_label('Mean AUC Score', rotation=270, labelpad=20, fontweight='bold')

# Highlight the absolute best parameter combination with a red star
plt.scatter(best_params['learning_rate'], best_params['max_depth'], 
            color='red', marker='*', s=400, label='Peak Performance')

plt.title('Continuous Search: Depth vs. Learning Rate', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Learning Rate (Continuous Range)', fontsize=12)
plt.ylabel('Max Tree Depth', fontsize=12)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Aggressive memory cleanup
del random_search
gc.collect()
```

Step 4: XGBoost Training & Probability of Default  (COMPUTED THE LR AND MAX DEPTH FROM STEP ABOVE RANDOM SEARCH

```python
# ═══════════════════════════════════════════════════════════
# XG BOOST
# ───────────────────────────────────────────────────────────


import xgboost as xgb
from sklearn.metrics import roc_auc_score
import gc
import re

print("1. Sanitizing column names for XGBoost")


def clean_column_names(df):
    df.columns = [re.sub(r'[\[\]<]', '_', str(col)) for col in df.columns]
    return df

# Apply the cleaning to both sets
X_train_encoded = clean_column_names(X_train_encoded)
X_test_encoded = clean_column_names(X_test_encoded)

print("2. Preparing DMatrix")
dtrain = xgb.DMatrix(X_train_encoded, label=y_train)
dtest = xgb.DMatrix(X_test_encoded, label=y_test)

# Calculate class imbalance ratio
ratio = (y_train == 0).sum() / (y_train == 1).sum()

print(f"3. Training XGBoost Model (Ratio: {ratio:.2f})")
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'scale_pos_weight': ratio,
    'learning_rate': 0.18309,
    'max_depth': 4,
    'tree_method': 'hist',
    'random_state': 42
}

model = xgb.train(params, dtrain, num_boost_round=100)

print("4. Generating Probability of Default (PD) scores")
test_pd = model.predict(dtest)

print("5. Evaluating Model Performance")
auc_score = roc_auc_score(y_test, test_pd)
gini = (2 * auc_score) - 1

print("\n" + "="*50)
print("STEP 4 COMPLETE: MODELING SUCCESS")
print(f"Out-of-Time ROC-AUC: {auc_score:.4f}")
print(f"Portfolio Gini Coefficient: {gini:.4f}")
print("="*50)

# Clear DMatrix to keep RAM fresh
del dtrain
gc.collect()
```

STEP 4 --- XG BOOST INTERNAL DIAGNOSTIC

STEP 5 --- PROFITABILITY TEST

```python
# ═══════════════════════════════════════════════════════════
# STEP 5: LIFETIME PROFITABILITY STRESS TEST
# ═══════════════════════════════════════════════════════════


import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("STEP 5: LIFETIME PROFITABILITY STRESS TEST")
print("=" * 60)

print("\n1. Assembling simulation dataframe")
test_financials = pd.DataFrame({
    'loan_amnt'    : X_test_encoded['loan_amnt'],
    'int_rate'     : test_df['int_rate'],
    'predicted_pd' : test_pd_calibrated,
})

test_financials['duration_years'] = np.where(
    test_df['term'].astype(str).str.contains('60'), 5, 3
)
test_financials['expected_loss'] = (
    test_financials['predicted_pd'] * test_financials['loan_amnt']
)

print("2. Calculating revenue scenarios...")

# Scenario A — market rates
base_rate = test_financials['int_rate'] / 100
test_financials['base_lifetime_revenue'] = (
    test_financials['loan_amnt'] * base_rate * test_financials['duration_years']
)

# Scenario B — 10% cap
cap_rate = np.minimum(test_financials['int_rate'], 10.0) / 100
test_financials['cap_lifetime_revenue'] = (
    test_financials['loan_amnt'] * cap_rate * test_financials['duration_years']
)

print("3. Applying OpEx as cost-to-income ratio")
DEFAULT_OPEX_RATE = 0.065  # 3.5% of revenue earned
test_financials['lifetime_opex'] = (
    test_financials['cap_lifetime_revenue'] * DEFAULT_OPEX_RATE
)

print("4. Computing net profits")
base_profit = (
    test_financials['base_lifetime_revenue']
    - test_financials['expected_loss']
    - (test_financials['base_lifetime_revenue'] * DEFAULT_OPEX_RATE)
)
cap_profit = (
    test_financials['cap_lifetime_revenue']
    - test_financials['expected_loss']
    - test_financials['lifetime_opex']
)

total_base_profit = base_profit.sum()
total_cap_profit  = cap_profit.sum()
revenue_destroyed = total_base_profit - total_cap_profit
cap_viable        = (cap_profit >= 0)
viable_count      = cap_viable.sum()
unviable_count    = (~cap_viable).sum()
unviable_pct      = unviable_count / len(cap_profit)

test_financials['base_profit'] = base_profit
test_financials['cap_profit']  = cap_profit

print("\n" + "=" * 60)
print("STEP 5 COMPLETE: LIFETIME PROFITABILITY STRESS TEST")
print("=" * 60)
print(f"  Total Test Set Loans : {len(test_financials):,}")
print("-" * 60)
print(f"  Scenario A (Baseline):  ${total_base_profit:>20,.2f}")
print(f"  Scenario B (10% Cap) :  ${total_cap_profit:>20,.2f}")
print("-" * 60)
print(f"  Revenue Wiped Out    :  ${revenue_destroyed:>20,.2f}")
print(f"  Loans Unviable       :  {unviable_count:,} ({unviable_pct:.1%})")
print("=" * 60)
```

```python
# ═══════════════════════════════════════════════════════════
#  LIVE FRED MACRO STRESS TEST  
# ═══════════════════════════════════════════════════════════

import datetime
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings
warnings.filterwarnings('ignore')

try:
    import pandas_datareader.data as web
    end_date   = datetime.date.today()
    start_date = end_date - datetime.timedelta(days=365)
    fred_data  = web.DataReader('FEDFUNDS', 'fred', start_date, end_date)
    FRED_RATE  = float(fred_data.iloc[-1, 0])
    print(f"Live FRED Rate secured: {FRED_RATE}%")
except Exception:
    FRED_RATE = 3.64
    print(f"FRED API unavailable — fallback: {FRED_RATE}%")

print("\n" + "=" * 60)
print("STEP 8: MACROECONOMIC STRESS TEST")
print("=" * 60)

def compute_scenario(opex_pct, fred_pct):
    opex_decimal = opex_pct / 100.0
    fred_decimal = fred_pct / 100.0          # ← FIX: /100 here

    fred_cost = (test_financials['loan_amnt']
                 * fred_decimal
                 * test_financials['duration_years'])
    opex_cost = test_financials['cap_lifetime_revenue'] * opex_decimal

    net_profit = (test_financials['cap_lifetime_revenue']
                  - test_financials['expected_loss']
                  - fred_cost
                  - opex_cost)
    viable = (net_profit >= 0)
    return net_profit.sum(), viable.sum(), (~viable).sum()

p1, v1, u1 = compute_scenario(opex_pct=0,   fred_pct=0)
p2, v2, u2 = compute_scenario(opex_pct=3.5, fred_pct=0)
p3, v3, u3 = compute_scenario(opex_pct=0,   fred_pct=FRED_RATE)
p4, v4, u4 = compute_scenario(opex_pct=3.5, fred_pct=FRED_RATE)

total = len(test_financials)
scenarios_data = [
    ("Cap Only\n(No Costs)",          p1, u1/total),
    ("Cap + OpEx\n(3.5%)",            p2, u2/total),
    (f"Cap + FRED\n({FRED_RATE}%)",   p3, u3/total),
    ("Fully Loaded\n(FRED + OpEx)",   p4, u4/total),
]

print(f"\n{'Scenario':<35} {'Net Profit':>15} {'% Unviable':>12}")
print("-" * 65)
for name, profit, u_pct in scenarios_data:
    print(f"  {name.replace(chr(10),' '):<33} ${profit:>14,.0f} {u_pct:>11.1%}")
print("=" * 65)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.patch.set_facecolor('#F8F9FA')
colors = ['#27AE60', '#E67E22', '#2980B9', '#C0392B']
names   = [s[0] for s in scenarios_data]
profits = [s[1] for s in scenarios_data]
unviables=[s[2] for s in scenarios_data]

ax1 = axes[0]
bars = ax1.bar(names, profits, color=colors,
               edgecolor='white', linewidth=1.5, width=0.6, zorder=3)
ax1.axhline(0, color='#2C3E50', linewidth=1.5, linestyle='--', zorder=4)
for bar, val in zip(bars, profits):
    sign = '+' if val >= 0 else ''
    ybuf = abs(max(profits, key=abs)) * 0.02
    ypos = val + ybuf if val >= 0 else val - ybuf * 3
    ax1.text(bar.get_x() + bar.get_width()/2, ypos,
             f"{sign}${val/1e6:.0f}M",
             ha='center', va='bottom' if val >= 0 else 'top',
             fontsize=11, fontweight='bold', color='#2C3E50')
ax1.set_title(f'Net Portfolio Profit by Scenario\n(10% Cap, FRED = {FRED_RATE}%)',
              fontsize=13, fontweight='bold', pad=15)
ax1.set_ylabel('Net Lifetime Profit (USD)', fontsize=11)
ax1.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f'${x/1e6:.0f}M'))
ax1.set_facecolor('#FAFAFA')
ax1.grid(axis='y', alpha=0.4, zorder=0)

ax2 = axes[1]
bars2 = ax2.bar(names, [u*100 for u in unviables],
                color=colors, edgecolor='white', linewidth=1.5,
                width=0.6, zorder=3)
for bar, val in zip(bars2, unviables):
    ax2.text(bar.get_x() + bar.get_width()/2, val*100 + 1.0,
             f"{val:.1%}", ha='center', va='bottom',
             fontsize=12, fontweight='bold', color='#2C3E50')
ax2.set_title('% Loans Rendered Unviable\n(Escalating Cost Layers)',
              fontsize=13, fontweight='bold', pad=15)
ax2.set_ylabel('% Unviable Loans', fontsize=11)
ax2.set_ylim(0, 105)
ax2.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f'{x:.0f}%'))
ax2.set_facecolor('#FAFAFA')
ax2.grid(axis='y', alpha=0.4, zorder=0)

plt.suptitle(
    f'Step 8: FRED Macro Stress Test  |  FRED = {FRED_RATE}%  |  '
    f'{total:,} Loans',
    fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('step8_fred_stress.png', dpi=150,
            bbox_inches='tight', facecolor='#F8F9FA')
plt.show()
```

```python
# ═══════════════════════════════════════════════════════════
#  INTERACTIVE DUAL-SLIDER MACRO SIMULATOR
# ═══════════════════════════════════════════════════════════

from ipywidgets import interact, FloatSlider
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

print("=" * 60)
print("STEP 9a: INTERACTIVE MACRO SIMULATOR")
print("=" * 60)

def interactive_stress_test(fred_rate, opex_rate_pct):
    fred_decimal = fred_rate     / 100.0    # ← FIX: /100 always
    opex_decimal = opex_rate_pct / 100.0

    fred_cost  = (test_financials['loan_amnt']
                  * fred_decimal
                  * test_financials['duration_years'])
    opex_cost  = test_financials['cap_lifetime_revenue'] * opex_decimal
    net_profit = (test_financials['cap_lifetime_revenue']
                  - test_financials['expected_loss']
                  - fred_cost - opex_cost)

    total_profit   = net_profit.sum()
    viable         = (net_profit >= 0)
    viable_count   = viable.sum()
    unviable_count = (~viable).sum()
    v_pct          = viable_count   / len(net_profit)
    u_pct          = unviable_count / len(net_profit)

    avg_rev  = test_financials['cap_lifetime_revenue'].mean()
    avg_el   = test_financials['expected_loss'].mean()
    avg_fred = fred_cost.mean()
    avg_opex = opex_cost.mean()
    avg_net  = avg_rev - avg_el - avg_fred - avg_opex

    sns.set_theme(style="whitegrid")
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))
    fig.patch.set_facecolor('#F8F9FA')

    # Panel A: Profit bar
    ax1 = axes[0]
    color = '#27AE60' if total_profit >= 0 else '#C0392B'
    ax1.bar(['Net Portfolio\nProfit'], [total_profit],
            color=color, edgecolor='white', linewidth=2, width=0.4)
    ax1.axhline(0, color='#2C3E50', linewidth=1.5, linestyle='--')
    sign = '+' if total_profit >= 0 else ''
    ax1.text(0, total_profit * 0.5,
             f"{sign}${total_profit/1e6:.1f}M",
             ha='center', fontsize=16, fontweight='bold', color='white')
    ax1.set_title(f'Net Profit\nFRED={fred_rate}% | OpEx={opex_rate_pct}%',
                  fontsize=12, fontweight='bold')
    ax1.yaxis.set_major_formatter(
        ticker.FuncFormatter(lambda x, _: f'${x/1e6:.0f}M'))
    ax1.set_facecolor('#FAFAFA')

    # Panel B: Viability pie
    ax2 = axes[1]
    wedges, _ = ax2.pie(
        [v_pct, u_pct],
        colors=['#27AE60', '#C0392B'],
        explode=(0.03, 0.03), startangle=90,
        wedgeprops=dict(linewidth=2, edgecolor='white')
    )
    ax2.legend(wedges,
               [f'Viable: {v_pct:.1%} ({viable_count:,})',
                f'Unviable: {u_pct:.1%} ({unviable_count:,})'],
               loc='lower center', bbox_to_anchor=(0.5, -0.15),
               fontsize=10)
    ax2.set_title('Loan Viability Under Cap',
                  fontsize=12, fontweight='bold')

    # Panel C: Per-loan cost breakdown
    ax3 = axes[2]
    categories = ['Cap Revenue', 'Exp. Loss', 'FRED Cost', 'OpEx', 'Net Profit']
    values     = [avg_rev, -avg_el, -avg_fred, -avg_opex, avg_net]
    bar_colors = ['#2980B9','#E74C3C','#E67E22','#9B59B6',
                  '#27AE60' if avg_net >= 0 else '#C0392B']
    ax3.bar(categories, values, color=bar_colors,
            edgecolor='white', linewidth=1.5)
    ax3.axhline(0, color='#2C3E50', linewidth=1, linestyle='--')
    buf = max(abs(v) for v in values) * 0.02
    for i, val in enumerate(values):
        ax3.text(i, val + (buf if val >= 0 else -buf),
                 f'${val:,.0f}', ha='center',
                 va='bottom' if val >= 0 else 'top', fontsize=9)
    ax3.set_title('Avg Per-Loan Cost Breakdown',
                  fontsize=12, fontweight='bold')
    ax3.set_ylabel('USD per Loan')
    ax3.set_facecolor('#FAFAFA')
    ax3.tick_params(axis='x', rotation=20)

    plt.suptitle(
        f'Macro Simulator — FRED={fred_rate}% | OpEx={opex_rate_pct}%',
        fontsize=13, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    print(f"\nSCENARIO: {fred_rate}% FRED | {opex_rate_pct}% OpEx")
    print("=" * 55)
    print(f"PROJECTED PROFIT  : ${total_profit:>15,.2f}")
    print(f"LOANS VIABLE      : {viable_count:>10,}  ({v_pct:.1%})")
    print(f"LOANS UNVIABLE    : {unviable_count:>10,}  ({u_pct:.1%})")
    print(f"AVG NET PER LOAN  : ${avg_net:>15,.2f}")
    print("=" * 55)

interact(
    interactive_stress_test,
    fred_rate=FloatSlider(
        value=FRED_RATE, min=0.0, max=15.0, step=0.125,
        description='FRED Rate %:', continuous_update=False,
        style={'description_width': 'initial'},
        layout={'width': '500px'}
    ),
    opex_rate_pct=FloatSlider(
        value=3.5, min=0.0, max=10.0, step=0.5,
        description='OpEx Rate %:', continuous_update=False,
        style={'description_width': 'initial'},
        layout={'width': '500px'}
    )
)
```

STEP 9 (b) --- MACRO SIMULATOR WITH FEATURE CHARACTERISTICS

```python
# ═══════════════════════════════════════════════════════════
# MACRO SIMULATOR WITH FEATURE CHARACTERISTICS
# ═══════════════════════════════════════════════════════════


from ipywidgets import interact, FloatSlider
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import pandas as pd
import numpy as np

print("=" * 60)
print("STEP 9b: INTERACTIVE MACRO SIMULATOR WITH BORROWER PROFILE")
print("=" * 60)

# ----------------------------------------------------------------
# FEATURE TYPE CLASSIFICATION
# Binary/dummy  -> Mode   (most common 0 or 1)
# Count data    -> Median (skew-resistant for whole numbers)
# Continuous    -> Mean   (best for ratios, amounts, scores)
# ----------------------------------------------------------------

BINARY_FEATURES = [
    'term_ 60 months',
    'home_ownership_RENT',
    'home_ownership_MORTGAGE',
    'pub_rec_bankruptcies',
    'pub_rec',
    'delinq_2yrs',
]

COUNT_FEATURES = [
    'open_acc',
    'total_acc',
    'acc_open_past_24mths',
    'mort_acc',
    'inq_last_6mths',
]

CONTINUOUS_FEATURES = [
    'interest_cost_ratio',
    'dti',
    'fico_range_low',
    'annual_inc',
    'loan_to_income',
    'revol_utilisation_rate',
    'payment_burden',
    'revol_bal',
    'emp_length',
]

# All 20 SHAP features in display order
SHAP_FEATURES = [
    'interest_cost_ratio', 'dti', 'acc_open_past_24mths', 'fico_range_low',
    'delinq_2yrs', 'annual_inc', 'loan_to_income', 'revol_utilisation_rate',
    'payment_burden', 'inq_last_6mths', 'open_acc', 'total_acc',
    'revol_bal', 'mort_acc', 'pub_rec', 'pub_rec_bankruptcies',
    'term_ 60 months', 'home_ownership_RENT', 'home_ownership_MORTGAGE',
    'emp_length'
]

# Keep only features that actually exist in X_test_encoded
SHAP_FEATURES = [f for f in SHAP_FEATURES if f in X_test_encoded.columns]


def stat_label(f):
    """Return the stat method label for display next to feature name."""
    if f in BINARY_FEATURES:
        return 'mode'
    if f in COUNT_FEATURES:
        return 'mdn'
    return 'avg'


# ----------------------------------------------------------------
# PROFILE FUNCTION
# Returns one representative value per feature using the correct
# statistic for that feature type.
# ----------------------------------------------------------------

def compute_profile(df_subset):
    if len(df_subset) == 0:
        return pd.Series({f: 0 for f in SHAP_FEATURES})

    profile = {}
    for f in SHAP_FEATURES:
        if f not in df_subset.columns:
            profile[f] = 0
        elif f in BINARY_FEATURES:
            mode_vals  = df_subset[f].mode()
            profile[f] = float(mode_vals.iloc[0]) if len(mode_vals) > 0 else 0
        elif f in COUNT_FEATURES:
            profile[f] = float(df_subset[f].median())
        else:
            profile[f] = float(df_subset[f].mean())

    return pd.Series(profile)


# ----------------------------------------------------------------
# BUILD MASTER SIMULATION DATAFRAME (once, outside the slider)
# Merges test_financials with all 20 SHAP feature columns
# ----------------------------------------------------------------

sim_master = test_financials.copy()
for col in SHAP_FEATURES:
    sim_master[col] = X_test_encoded[col].values

print("  SHAP features loaded  : {}".format(len(SHAP_FEATURES)))
print("  Master sim shape      : {}".format(sim_master.shape))
print("  Default FRED rate     : {}%".format(FRED_RATE))
print("")
print("  Stat method per feature type:")
print("    Binary  ({}) -> MODE".format(len([f for f in BINARY_FEATURES if f in SHAP_FEATURES])))
print("    Count   ({}) -> MEDIAN".format(len([f for f in COUNT_FEATURES if f in SHAP_FEATURES])))
print("    Continuous -> MEAN")
print("")
print("SIMULATOR READY - use sliders below")
print("")


# ----------------------------------------------------------------
# SIMULATOR FUNCTION
# ----------------------------------------------------------------

def interactive_stress_test(fred_rate, opex_rate_pct):

    fred_decimal = fred_rate     / 100.0
    opex_decimal = opex_rate_pct / 100.0

    # Per-loan net profit calculation
    fred_cost  = (sim_master['loan_amnt']
                  * fred_decimal
                  * sim_master['duration_years'])
    opex_cost  = sim_master['cap_lifetime_revenue'] * opex_decimal
    net_profit = (sim_master['cap_lifetime_revenue']
                  - sim_master['expected_loss']
                  - fred_cost
                  - opex_cost)

    total_profit   = net_profit.sum()
    viable_mask    = (net_profit >= 0)
    viable_count   = viable_mask.sum()
    unviable_count = (~viable_mask).sum()
    v_pct          = viable_count   / len(net_profit)
    u_pct          = unviable_count / len(net_profit)

    # Cost breakdown averages
    avg_rev  = sim_master['cap_lifetime_revenue'].mean()
    avg_el   = sim_master['expected_loss'].mean()
    avg_fred = fred_cost.mean()
    avg_opex = opex_cost.mean()
    avg_net  = avg_rev - avg_el - avg_fred - avg_opex

    # Borrower profiles using correct stat per feature type
    viable_loans   = sim_master[viable_mask]
    unviable_loans = sim_master[~viable_mask]

    profile_viable   = compute_profile(viable_loans)
    profile_unviable = compute_profile(unviable_loans)
    profile_diff     = profile_viable - profile_unviable

    # ----------------------------------------------------------------
    # DASHBOARD: 2 rows x 3 cols
    # Row 1: Profit bar  | Viability pie   | Cost waterfall
    # Row 2: Viable prof | Unviable prof   | Difference bar
    # ----------------------------------------------------------------
    sns.set_theme(style="whitegrid")
    fig = plt.figure(figsize=(22, 14))
    fig.patch.set_facecolor('#F8F9FA')

    gs  = fig.add_gridspec(2, 3, hspace=0.48, wspace=0.38)
    ax1 = fig.add_subplot(gs[0, 0])
    ax2 = fig.add_subplot(gs[0, 1])
    ax3 = fig.add_subplot(gs[0, 2])
    ax4 = fig.add_subplot(gs[1, 0])
    ax5 = fig.add_subplot(gs[1, 1])
    ax6 = fig.add_subplot(gs[1, 2])

    # Panel 1: Net Profit Bar
    p_color = '#27AE60' if total_profit >= 0 else '#C0392B'
    ax1.bar(['Net Portfolio\nProfit'], [total_profit],
            color=p_color, edgecolor='white', linewidth=2, width=0.45)
    ax1.axhline(0, color='#2C3E50', linewidth=1.5, linestyle='--')
    sign  = '+' if total_profit >= 0 else ''
    lbl_y = total_profit * 0.5 if total_profit != 0 else 0
    ax1.text(0, lbl_y,
             "{}${:.1f}M".format(sign, total_profit / 1e6),
             ha='center', fontsize=16, fontweight='bold', color='white')
    ax1.set_title(
        "Net Profit\nFRED = {}%  |  OpEx = {}%".format(fred_rate, opex_rate_pct),
        fontsize=12, fontweight='bold')
    ax1.yaxis.set_major_formatter(
        ticker.FuncFormatter(lambda x, _: '${:.0f}M'.format(x / 1e6)))
    ax1.set_facecolor('#FAFAFA')

    # Panel 2: Viability Pie
    if viable_count > 0 and unviable_count > 0:
        wedges, _ = ax2.pie(
            [v_pct, u_pct],
            colors=['#27AE60', '#C0392B'],
            explode=(0.03, 0.03),
            startangle=90,
            wedgeprops=dict(linewidth=2, edgecolor='white')
        )
        ax2.legend(
            wedges,
            [
                "Viable: {:.1%}  ({:,} loans)".format(v_pct, viable_count),
                "Unviable: {:.1%}  ({:,} loans)".format(u_pct, unviable_count)
            ],
            loc='lower center',
            bbox_to_anchor=(0.5, -0.16),
            fontsize=10
        )
    elif viable_count == 0:
        ax2.text(0, 0, 'All Loans\nUnviable',
                 ha='center', va='center',
                 fontsize=14, fontweight='bold', color='#C0392B')
    else:
        ax2.text(0, 0, 'All Loans\nViable',
                 ha='center', va='center',
                 fontsize=14, fontweight='bold', color='#27AE60')
    ax2.set_title('Loan Viability Under 10% Cap', fontsize=12, fontweight='bold')

    # Panel 3: Per-Loan Cost Waterfall
    cat   = ['Cap\nRevenue', 'Exp.\nLoss', 'FRED\nCost', 'OpEx', 'Net\nProfit']
    vals  = [avg_rev, -avg_el, -avg_fred, -avg_opex, avg_net]
    bcols = ['#2980B9', '#E74C3C', '#E67E22', '#9B59B6',
             '#27AE60' if avg_net >= 0 else '#C0392B']
    ax3.bar(cat, vals, color=bcols, edgecolor='white', linewidth=1.5)
    ax3.axhline(0, color='#2C3E50', linewidth=1, linestyle='--')
    buf = max(abs(v) for v in vals) * 0.03
    for i, val in enumerate(vals):
        ax3.text(i, val + (buf if val >= 0 else -buf),
                 '${:,.0f}'.format(val),
                 ha='center',
                 va='bottom' if val >= 0 else 'top',
                 fontsize=9, fontweight='bold')
    ax3.set_title('Avg Per-Loan Cost Breakdown', fontsize=12, fontweight='bold')
    ax3.set_ylabel('USD per Loan')
    ax3.set_facecolor('#FAFAFA')

    # Shared y-axis labels for panels 4-6
    y_pos       = np.arange(len(SHAP_FEATURES))
    feat_labels = ['{:<20} [{}]'.format(f[:20], stat_label(f))
                   for f in SHAP_FEATURES]

    # Panel 4: Viable Borrower Profile
    ax4.barh(y_pos, profile_viable.values,
             color='#27AE60', edgecolor='white', linewidth=0.8, alpha=0.85)
    ax4.set_yticks(y_pos)
    ax4.set_yticklabels(feat_labels, fontsize=8.5)
    ax4.set_title(
        "Viable Borrower Profile  (n = {:,})\n"
        "[avg]=mean  [mdn]=median  [mode]=mode".format(viable_count),
        fontsize=11, fontweight='bold', color='#1E8449')
    ax4.set_xlabel('Representative Feature Value', fontsize=9)
    ax4.set_facecolor('#F0FFF4')
    ax4.axvline(0, color='#2C3E50', linewidth=0.8, linestyle='--')
    max_v = max(abs(v) for v in profile_viable.values) if len(profile_viable) > 0 else 1
    for i, val in enumerate(profile_viable.values):
        offset = max_v * 0.01
        ax4.text(
            val + (offset if val >= 0 else -offset), i,
            '{:.2f}'.format(val),
            va='center',
            ha='left' if val >= 0 else 'right',
            fontsize=7.5, color='#1A5276')

    # Panel 5: Unviable Borrower Profile
    ax5.barh(y_pos, profile_unviable.values,
             color='#C0392B', edgecolor='white', linewidth=0.8, alpha=0.85)
    ax5.set_yticks(y_pos)
    ax5.set_yticklabels(feat_labels, fontsize=8.5)
    ax5.set_title(
        "Unviable Borrower Profile  (n = {:,})\n"
        "[avg]=mean  [mdn]=median  [mode]=mode".format(unviable_count),
        fontsize=11, fontweight='bold', color='#922B21')
    ax5.set_xlabel('Representative Feature Value', fontsize=9)
    ax5.set_facecolor('#FFF5F5')
    ax5.axvline(0, color='#2C3E50', linewidth=0.8, linestyle='--')
    max_u = max(abs(v) for v in profile_unviable.values) if len(profile_unviable) > 0 else 1
    for i, val in enumerate(profile_unviable.values):
        offset = max_u * 0.01
        ax5.text(
            val + (offset if val >= 0 else -offset), i,
            '{:.2f}'.format(val),
            va='center',
            ha='left' if val >= 0 else 'right',
            fontsize=7.5, color='#922B21')

    # Panel 6: Difference (Viable - Unviable)
    diff_colors = ['#27AE60' if v >= 0 else '#C0392B'
                   for v in profile_diff.values]
    ax6.barh(y_pos, profile_diff.values,
             color=diff_colors, edgecolor='white', linewidth=0.8, alpha=0.85)
    ax6.set_yticks(y_pos)
    ax6.set_yticklabels(feat_labels, fontsize=8.5)
    ax6.set_title(
        "Difference: Viable minus Unviable\n"
        "Green = viable score higher  |  Red = unviable score higher",
        fontsize=11, fontweight='bold')
    ax6.set_xlabel('Difference in Representative Value', fontsize=9)
    ax6.set_facecolor('#FAFAFA')
    ax6.axvline(0, color='#2C3E50', linewidth=1.2, linestyle='--')
    diff_vals = list(profile_diff.values)
    max_d = max(abs(v) for v in diff_vals) if any(diff_vals) else 0.01
    for i, val in enumerate(diff_vals):
        offset = max_d * 0.01
        ax6.text(
            val + (offset if val >= 0 else -offset), i,
            '{:+.2f}'.format(val),
            va='center',
            ha='left' if val >= 0 else 'right',
            fontsize=7.5,
            color='#1E8449' if val >= 0 else '#922B21')

    plt.suptitle(
        "Macro Simulator  |  FRED = {}%  |  OpEx = {}%  |  "
        "Viable: {:.1%} ({:,})  |  Unviable: {:.1%} ({:,})".format(
            fred_rate, opex_rate_pct,
            v_pct, viable_count,
            u_pct, unviable_count),
        fontsize=13, fontweight='bold', y=1.01)

    plt.savefig(
        'step9_fred{}_opex{}.png'.format(fred_rate, opex_rate_pct),
        dpi=120, bbox_inches='tight', facecolor='#F8F9FA')
    plt.show()

    # ----------------------------------------------------------------
    # CONSOLE: Full numerical profile table
    # ----------------------------------------------------------------
    print("")
    print("SCENARIO: {}% FRED  |  {}% OpEx".format(fred_rate, opex_rate_pct))
    print("=" * 72)
    print("  PROJECTED PROFIT   : ${:>15,.2f}".format(total_profit))
    print("  LOANS VIABLE       : {:>10,}  ({:.1%})".format(viable_count, v_pct))
    print("  LOANS UNVIABLE     : {:>10,}  ({:.1%})".format(unviable_count, u_pct))
    print("  AVG NET PER LOAN   : ${:>15,.2f}".format(avg_net))
    print("-" * 72)
    print("  {:<26} {:<6} {:>10}  {:>10}  {:>10}".format(
        'FEATURE', 'STAT', 'VIABLE', 'UNVIABLE', 'DIFF'))
    print("-" * 72)

    for f in SHAP_FEATURES:
        stat  = stat_label(f)
        v_val = profile_viable.get(f, 0)
        u_val = profile_unviable.get(f, 0)
        diff  = v_val - u_val
        if diff > 0:
            arrow = '(+)'
        elif diff < 0:
            arrow = '(-)'
        else:
            arrow = '(=)'
        print("  {:<26} [{:<4}] {:>10.3f}  {:>10.3f}  {:>+9.3f} {}".format(
            f[:26], stat, v_val, u_val, diff, arrow))

    print("=" * 72)
    print("  (+) = viable borrowers score HIGHER on this feature")
    print("  (-) = viable borrowers score LOWER  on this feature")
    print("=" * 72)


# ----------------------------------------------------------------
# RENDER SLIDERS
# ----------------------------------------------------------------
interact(
    interactive_stress_test,
    fred_rate=FloatSlider(
        value=FRED_RATE,
        min=0.0,
        max=15.0,
        step=0.125,
        description='FRED Rate %:',
        continuous_update=False,
        style={'description_width': 'initial'},
        layout={'width': '500px'}
    ),
    opex_rate_pct=FloatSlider(
        value=3.5,
        min=0.0,
        max=10.0,
        step=0.5,
        description='OpEx Rate %:',
        continuous_update=False,
        style={'description_width': 'initial'},
        layout={'width': '500px'}
    )
)
```

STEP 10 --- CONFUSION MATRIX

```python
# ═══════════════════════════════════════════════════════════
# CONFUSION MATRIX & FULL MODEL EVALUATION DASHBOARD
# ═══════════════════════════════════════════════════════════

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import (confusion_matrix, classification_report,
                             roc_curve, roc_auc_score)

print("=" * 60)
print("CONFUSION MATRIX & MODEL EVALUATION")
print("=" * 60)

print("\n[1/4] Finding optimal threshold (Youden's J)")
fpr, tpr, thresholds = roc_curve(y_test, test_pd)
optimal_idx       = (tpr - fpr).argmax()
optimal_threshold = thresholds[optimal_idx]
print(f"      Naive threshold   : 0.500")
print(f"      Optimal threshold : {optimal_threshold:.3f}")

print("\n[2/4] Generating predictions")
y_pred_optimal = (test_pd >= optimal_threshold).astype(int)
y_pred_naive   = (test_pd >= 0.50).astype(int)

print("\n[3/4] Computing metrics")
cm             = confusion_matrix(y_test, y_pred_optimal)
tn, fp, fn, tp = cm.ravel()
total          = tn + fp + fn + tp
precision      = tp / (tp + fp)
recall         = tp / (tp + fn)
specificity    = tn / (tn + fp)
f1             = 2 * precision * recall / (precision + recall)
auc_score      = roc_auc_score(y_test, test_pd)
gini           = 2 * auc_score - 1

print("\n[4/4] Rendering 3-panel dashboard...")
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.patch.set_facecolor('#F8F9FA')

# Panel A: Confusion Matrix
ax = axes[0]
labels = np.array([
    [f"True Negative\n{tn:,}", f"False Positive\n{fp:,}"],
    [f"False Negative\n{fn:,}", f"True Positive\n{tp:,}"]
])
sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', ax=ax,
            linewidths=2, linecolor='white', cbar=False,
            annot_kws={"size": 13, "weight": "bold"})
notes = [['Safe loan\ncorrectly cleared', 'Safe loan\nwrongly flagged'],
         ['Missed default\n(credit loss!)',  'Default\ncorrectly caught ✓']]
for i in range(2):
    for j in range(2):
        ax.text(j+0.5, i+0.75, notes[i][j],
                ha='center', va='center',
                fontsize=8, color='#555', style='italic')
ax.set_title(f'Confusion Matrix\n(Threshold = {optimal_threshold:.3f})',
             fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel('Predicted Label', fontsize=11, fontweight='bold')
ax.set_ylabel('Actual Label',    fontsize=11, fontweight='bold')
ax.set_xticklabels(['Safe (0)', 'Default (1)'], fontsize=11)
ax.set_yticklabels(['Safe (0)', 'Default (1)'], fontsize=11, rotation=0)

# Panel B: ROC Curve
ax2 = axes[1]
ax2.plot(fpr, tpr, color='#1E2761', lw=2.5,
         label=f'XGBoost  AUC = {auc_score:.4f}')
ax2.fill_between(fpr, tpr, alpha=0.08, color='#1E2761')
ax2.plot([0,1],[0,1], 'k--', lw=1.2, label='Random  AUC = 0.50')
ax2.scatter(fpr[optimal_idx], tpr[optimal_idx],
            color='#F39C12', s=150, zorder=5,
            label=f'Optimal cut-off ({optimal_threshold:.3f})')
ax2.axvline(fpr[optimal_idx], color='#F39C12', lw=1, linestyle=':', alpha=0.6)
ax2.axhline(tpr[optimal_idx], color='#F39C12', lw=1, linestyle=':', alpha=0.6)
ax2.set_title('ROC Curve\n(Out-of-Time Test 2016–2018)',
              fontsize=13, fontweight='bold', pad=15)
ax2.set_xlabel('False Positive Rate (1 − Specificity)', fontsize=11)
ax2.set_ylabel('True Positive Rate (Sensitivity)',       fontsize=11)
ax2.legend(loc='lower right', fontsize=10)
ax2.set_xlim([0,1]); ax2.set_ylim([0,1.02])
ax2.set_facecolor('#FAFAFA')
ax2.grid(True, alpha=0.3)
ax2.text(0.55, 0.15,
         f"Gini: {gini:.4f}\nBasel II floor: 0.30\n"
         f"Day-1 benchmark: >0.35\n✓ Model exceeds both",
         fontsize=9.5, color='#1E2761',
         bbox=dict(boxstyle='round,pad=0.5',
                   facecolor='#EEF2F9', edgecolor='#CADCFC'))

# Panel C: Threshold comparison table
ax3 = axes[2]
ax3.axis('off')
rep_opt  = classification_report(y_test, y_pred_optimal, output_dict=True)
rep_naive= classification_report(y_test, y_pred_naive,   output_dict=True)
table_data = [
    ["Metric",            "Threshold 0.50\n(Naive)",
                          f"Threshold {optimal_threshold:.3f}\n(Optimal)"],
    ["Accuracy",          f"{rep_naive['accuracy']:.3f}",
                          f"{rep_opt['accuracy']:.3f}"],
    ["Default Precision", f"{rep_naive['1']['precision']:.3f}",
                          f"{rep_opt['1']['precision']:.3f}"],
    ["Default Recall",    f"{rep_naive['1']['recall']:.3f}",
                          f"{rep_opt['1']['recall']:.3f}"],
    ["Default F1",        f"{rep_naive['1']['f1-score']:.3f}",
                          f"{rep_opt['1']['f1-score']:.3f}"],
    ["Safe Recall",       f"{rep_naive['0']['recall']:.3f}",
                          f"{rep_opt['0']['recall']:.3f}"],
    ["ROC-AUC",           f"{auc_score:.4f}", f"{auc_score:.4f}"],
    ["Gini",              f"{gini:.4f}",      f"{gini:.4f}"],
]
tbl = ax3.table(cellText=table_data[1:], colLabels=table_data[0],
                cellLoc='center', loc='center',
                bbox=[0, 0.22, 1, 0.72])
tbl.auto_set_font_size(False)
tbl.set_fontsize(11)
for j in range(3):
    tbl[(0,j)].set_facecolor('#1E2761')
    tbl[(0,j)].set_text_props(color='white', fontweight='bold')
for i in range(1, len(table_data)):
    tbl[(i,0)].set_facecolor('#EEF2F9')
    tbl[(i,0)].set_text_props(fontweight='bold')
    tbl[(i,2)].set_facecolor('#E8F4FD')
ax3.set_title('Threshold Comparison\nNaive vs Optimal',
              fontsize=13, fontweight='bold', pad=15)
ax3.text(0.5, 0.1,
         f"MODEL SUMMARY\n"
         f"─────────────────────────\n"
         f"Test loans :  {total:>10,}\n"
         f"True Pos   :  {tp:>10,}\n"
         f"False Neg  :  {fn:>10,}  ← missed\n"
         f"Sensitivity:  {recall:>9.1%}\n"
         f"Specificity:  {specificity:>9.1%}\n"
         f"F1 Score   :  {f1:>9.4f}",
         transform=ax3.transAxes, ha='center', va='center',
         fontsize=10, fontfamily='monospace',
         bbox=dict(boxstyle='round,pad=0.6',
                   facecolor='#1E2761', edgecolor='none'),
         color='white')

plt.suptitle(
    'XGBoost Credit Default Model — Full Evaluation Dashboard\n'
    'Lending Club | Out-of-Time Test: 2016–2018',
    fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('confusion_matrix_dashboard.png', dpi=150,
            bbox_inches='tight', facecolor='#F8F9FA')
plt.show()

print("\n" + "=" * 60)
print("FINAL MODEL RESULTS")
print("=" * 60)
print(f"  ROC-AUC           : {auc_score:.4f}")
print(f"  Gini Coefficient  : {gini:.4f}  (Day-1 benchmark >0.35 ✓)")
print(f"  Optimal Threshold : {optimal_threshold:.3f}")
print(f"  Sensitivity       : {recall:.4f}")
print(f"  Specificity       : {specificity:.4f}")
print(f"  F1 Score          : {f1:.4f}")
print("-" * 60)
print(f"  True Positives    : {tp:>10,}  defaults correctly flagged")
print(f"  True Negatives    : {tn:>10,}  safe loans correctly cleared")
print(f"  False Positives   : {fp:>10,}  safe loans wrongly flagged")
print(f"  False Negatives   : {fn:>10,}  missed defaults")
print("=" * 60)
```
